## Задание 1

```python
ratings = pd.read_csv(r'C:\Users\79111\Notebooks\ml-latest-small\ratings.csv')
movies = pd.read_csv(r'C:\Users\79111\Notebooks\ml-latest-small\movies.csv')

merged = ratings.merge(movies, on='movieId', how='left')[['rating', 'title']]
```

```python
mean_scores = merged.groupby('title').mean().reset_index()
mean_scores.head()
```

```python
def classify(movie):
    if movie['rating'] <= 2:
        return 'low rating'
    if movie['rating'] <= 4:
        return 'average rating'
    return 'high rating'
```

```python
mean_scores['class'] = mean_scores.apply(classify, axis=1)
mean_scores.head(10)
```

## Задание 2

```python
keywords = pd.read_csv(r'C:\Users\79111\Notebooks\ml-latest-small\keywords.csv')

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_classifier(row):
    for key, value in geo_data.items():
        for city in value:
            if city in row['keyword']:
                return key
    return 'undefined'
```

```python
keywords['region'] = keywords.apply(geo_classifier, axis=1)
keywords[keywords['region'] != 'undefined'].head()
```

## Задание 3

```python
ratings = pd.read_csv(r'C:\Users\79111\Notebooks\ml-latest-small\ratings.csv')
movies = pd.read_csv(r'C:\Users\79111\Notebooks\ml-latest-small\movies.csv')

movies_with_rating = ratings.merge(movies, on='movieId', how='left')[['rating', 'title']]
```

```python
years = list(range(1950,2011))
```

```python
def production_year(movie):
    for year in years:
        if str(year) in movie['title']:
            return year
    return 1900
```

```python
movies_with_rating['year'] = movies_with_rating.apply(production_year, axis=1)
movies_with_rating.head()
```

```python
movies_with_rating.groupby('year').mean().sort_values(by='rating', ascending=False).head(30)
```
